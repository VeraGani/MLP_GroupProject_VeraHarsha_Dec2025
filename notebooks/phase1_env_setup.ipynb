```python
# Create a modified requirements file for Colab
new_lines = []
with open("requirements.txt") as f:
    for line in f:
        if line.strip().startswith("numpy=="):
            new_lines.append("numpy>=1.24.0\n")
        else:
            new_lines.append(line)

with open("requirements_colab.txt", "w") as f:
    f.writelines(new_lines)

print("Wrote requirements_colab.txt")
```
